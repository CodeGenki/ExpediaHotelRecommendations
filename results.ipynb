```
data = {
    'Model':['XG Boost','Naive Bayes','KNN',\
          'RF (multiclass)','RF (binary)',\
          'Logistic Regression','Baseline'
         ],
    'MAP@5 Score':[.25289,.02690,.12458,.15624,.28141,.10200,.07213]
}

import pandas as pd

df = pd.DataFrame(data)
df = df.sort_values(by=['MAP@5 Score'])
df.head(10)
```

```
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
fig, ax = plt.subplots()
fig.set_size_inches(26, 16)
ax = sns.barplot(x="Model", y="MAP@5 Score", data = df)
for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(25)
```
