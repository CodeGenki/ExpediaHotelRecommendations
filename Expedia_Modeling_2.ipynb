```
# Function to convert date object into relevant attributes
def convert_date_into_days(df):
    df['srch_ci'] = pd.to_datetime(df['srch_ci'], errors = 'coerce')
    df['srch_co'] = pd.to_datetime(df['srch_co'])
    df['date_time'] = pd.to_datetime(df['date_time'])
    
    df['stay_dur'] = (df['srch_co'] - df['srch_ci']).astype('timedelta64[D]')
    df['no_of_days_bet_booking'] = (df['srch_ci'] - df['date_time']).astype('timedelta64[D]')
    
    # For hotel check-in
    # Month, Year, Day
    df['Cin_day'] = df["srch_ci"].apply(lambda x: x.day)
    df['Cin_month'] = df["srch_ci"].apply(lambda x: x.month)
    df['Cin_year'] = df["srch_ci"].apply(lambda x: x.year)
```

```
import ml_metrics as metrics

def map5eval(target,predictions):
  pred = predictions.argsort(axis=1)[:,-np.arange(1,6)]
  return metrics.mapk(target,pred,k=5)
  
```

```
def map5eval(preds, actual):
    #actual = dtrain.get_label()
    predicted = preds.argsort(axis=1)[:,-np.arange(1,6)]
    metric = 0.
    for i in range(5):
        metric += np.sum(actual==predicted[:,i])/(i+1)
    metric /= actual.shape[0]
    return 'MAP@5', metric
```

```
import xgboost as xgb

xgb_params = {
    "objective": "multi:softmax",
    "eta": 0.3,
    "num_class": 100,
    "max_depth": 5,
    "feval": "map5eval",
    "n_estimators": 100
    #"silent": 1
}

xgb_model = xgb.XGBClassifier(**xgb_params)
xgb_model.fit(Xtrain,ytrain)
```
